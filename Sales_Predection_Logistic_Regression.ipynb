```python
import pandas as pd
```

```python
ds = pd.read_csv("DigitalAd_dataset.csv")
```

```python
ds.shape
```

```python
ds.describe()
```

```python
ds.isna().sum()
```

```python
x= ds.iloc[:,:-1].values
x
```

```python
y = ds.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=14)
```

```python
lr.fit(x_train, y_train)
```

```python
y_tr_pred = lr.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_train, y_tr_pred)
```

```python
accuracy_score(y_train, y_tr_pred)*100
```

```python
y_te_pred = lr.predict(x_test)
```

```python
confusion_matrix(y_test, y_te_pred )
```

```python
accuracy_score(y_test, y_te_pred)*100
```

```python
age = int(input("Enter the age: "))
sal = int(input("Ente the saleary: "))
new_cust = [[age, sal]]
result = lr.predict(scaler.transform(new_cust))
if result == 1 :
  print("Will Buy")
else:
  print("Will not buy")
```
